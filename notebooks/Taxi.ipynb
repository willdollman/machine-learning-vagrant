# This is a complete implementation!

```python
import numpy as np
import gym
import random
```

# Create the environment

```python
env = gym.make("Taxi-v2")
env.render()
```

```python
action_size = env.action_space.n
print("Action size: ", action_size)

state_size = env.observation_space.n
print("State size: ", state_size)
```

```python
q_table = np.zeros((state_size, action_size))
print(q_table)
```

```python
total_episodes = 50000    # total episodes
#total_episodes = 1
total_test_episodes = 100 # total test episodes
#total_test_episodes = 1
max_steps = 99            # max steps per episode
#max_steps = 9            # max steps per episode

learning_rate = 0.7       # learning rate
gamma = 0.618             # discounting rate

# Exploration parameters
epsilon = 1.0             # exploration rate
max_epsilon = 1.0         # exploration probability at start
min_epsilon = 0.01        # minimum exploration probability
decay_rate = 0.01         # exponential decay rate for exploration prob
```

# Step 4: Q learning algorithm

1. Initialise Q-values `Q(s,a)` arbitrarily for all state-action pairs
2. Continue until learning is stopped...
    3. Choose an action `(a)` in the current world state `(s)` based on current Q-value estimates `(Q(s,.))`
    4. Take the action `(a)` and observe the outcome state `(s')` and reward `(r)`
    5. Update Q-value of taking that action at that state (`Q(s,a)`) using te Bernman (?) equation - that's `Q(s,a) + ...`

```python
# 2: For life, or until learning is stopped
for episode in range(total_episodes):
    # Reset the environment
    state = env.reset()
    step = 0
    done = False
    
    for step in range(max_steps):
        # 3. Choose an action a in the current world state (s)
        ## First we randomise a number
        exp_exp_tradeoff = random.uniform(0,1)
        #print("Tradeoff: ", exp_exp_tradeoff)
        
        # If this number > greater than epsilon, then exploitation 
        # (take the biggest Q-value for this state)
        if exp_exp_tradeoff > epsilon:
            action = np.argmax(q_table[state,:])
            #print("Exploitation, using action ", action)
            
        # Else perform a random choice, then exploration
        else:
            action = env.action_space.sample()
            #print("Exploration, using random action", action)
            
        # Take action (a) and observe the outcome state (s') and reward (r)
        new_state, reward, done, info = env.step(action)
        
        # Update Q(s,a) := Q(s,a) + lr [R(s,a) + gamma * max Q(s', a') - Q(s,a)]
        q_table[state, action] = q_table[state, action] + learning_rate * ( reward + gamma * np.max(q_table[new_state, :]) - q_table[state, action] )
        
        # Our new state is state
        state = new_state
        
        if done == True:
            break
            
    print("Completed episode", episode)
    episode += 1
    
    # Reduce epsilon (because we need less and less exploration)
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
```

# Use our Q-table to play Taxi!

* After running 50,000 training episodes, we can use the Q-table as a cheat sheet to play Taxi

```python
env.reset()
rewards = []

for episode in range(total_test_episodes):
    state = env.reset()
    step = 0
    done = False
    total_rewards = 0
    print("*********************")
    print("Episode ", episode)
    
    for step in range(max_steps):
        env.render()
        # Take the action (index) that has the max expected future reward given that state
        action = np.argmax(q_table[state,:])
        
        new_state, reward, done, info = env.step(action)
        
        total_rewards += reward
        
        if done:
            rewards.append(total_rewards)
            print("Score: ", total_rewards)
            break
        state = new_state
env.close()
print("Score over time: " + str(sum(rewards)/total_test_episodes))
```
